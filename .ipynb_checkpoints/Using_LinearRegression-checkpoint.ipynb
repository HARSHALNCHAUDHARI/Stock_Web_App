```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense
import os

# Fetch stock data
start = '2000-01-01'
end = '2025-01-01'
df = yf.download('AAPL', start=start, end=end)

# Preprocessing the data
scaler = MinMaxScaler(feature_range=(0, 1))
data_training = pd.DataFrame(df['Close'][0:int(len(df)*0.70)])
data_testing = pd.DataFrame(df['Close'][int(len(df)*0.70):int(len(df))])

data_training_array = scaler.fit_transform(data_training)

# Creating the training dataset for Linear Regression
x_train = []
y_train = []
for i in range(100, data_training_array.shape[0]): 
    x_train.append(data_training_array[i-100:i, 0])  # Only using the closing price
    y_train.append(data_training_array[i, 0])

x_train, y_train = np.array(x_train), np.array(y_train)

# Reshaping x_train to be 2D (samples, features)
x_train = x_train.reshape(x_train.shape[0], x_train.shape[1])

# Linear Regression Model
lr_model = LinearRegression()
lr_model.fit(x_train, y_train)

# Save the Linear Regression model using Keras (as .keras file)
# Wrapping the model in Keras for saving
keras_model = Sequential()
keras_model.add(Dense(units=1, input_dim=x_train.shape[1]))

# Reshape the coefficients properly to match the Dense layer
coefficients = lr_model.coef_.reshape(x_train.shape[1], 1)
intercept = lr_model.intercept_.reshape(1,)

# Set the weights and biases
keras_model.set_weights([coefficients, intercept])

# Save the model to a .keras file
model_file_path = 'linear_regression_model.keras'
keras_model.save(model_file_path)

print(f"Model saved as {model_file_path}")

# Prepare test data for prediction
data_testing_array = scaler.transform(data_testing)

# Create the test dataset for Linear Regression
x_test = []
y_test = []
for i in range(100, data_testing_array.shape[0]):
    x_test.append(data_testing_array[i-100:i, 0])  # Only using the closing price
    y_test.append(data_testing_array[i, 0])

x_test = np.array(x_test)

# Reshaping x_test to be 2D (samples, features)
x_test = x_test.reshape(x_test.shape[0], x_test.shape[1])

# Make predictions
y_predicted_lr = lr_model.predict(x_test)

# Inverse scaling the predictions for Linear Regression
scale_factor = 1 / scaler.scale_[0]
y_predicted_lr = y_predicted_lr * scale_factor
y_test = np.array(y_test) * scale_factor

# Plotting the results for Linear Regression predictions
plt.figure(figsize=(12, 6))
plt.plot(y_test, label='Original Price', color='blue')
plt.plot(y_predicted_lr, label='Predicted Price (Linear Regression)', color='green')
plt.title('Original vs Predicted Prices - Linear Regression')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()

# Save the plot
plot_file_path = 'static/linear_regression_plot.png'
plt.savefig(plot_file_path)
```
